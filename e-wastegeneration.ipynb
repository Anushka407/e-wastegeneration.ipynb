```python
# Import necessary libraries
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Define paths to the dataset
trainpath = r'C:\Users\lenovo\Downloads\e-waste\modified-dataset\train'
validpath = r'C:\Users\lenovo\Downloads\e-waste\modified-dataset\val'
testpath = r'C:\Users\lenovo\Downloads\e-waste\modified-dataset\test'

# Load training data
datatrain = tf.keras.utils.image_dataset_from_directory(
    trainpath,
    shuffle=True,
    image_size=(128, 128),
    batch_size=32,
    validation_split=False
)

# Load validation data
datavalid = tf.keras.utils.image_dataset_from_directory(
    validpath,
    shuffle=True,
    image_size=(128, 128),
    batch_size=32,
    validation_split=False
)

# Load testing data
datatest = tf.keras.utils.image_dataset_from_directory(
    testpath,
    shuffle=False,  # Keep shuffle=False for evaluation
    image_size=(128, 128),
    batch_size=32,
    validation_split=False
)

# Check class names
class_names = datatrain.class_names
print(f"Total Classes: {len(class_names)}")
print(f"Class Names: {class_names}")
```

```python
plt.figure(figsize=(12, 8))
for images, labels in datatrain.take(1):
    for i in range(9):
        plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[labels[i]])
        plt.axis("off")
plt.show()
```
